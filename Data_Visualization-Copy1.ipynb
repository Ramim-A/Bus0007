**1. Relationship between the total bill and the tip amount:**

You can use a scatter plot to visualize the relationship between the total bill and the tip amount. This will help you understand the correlation between these two variables and the extent to which people tip based on the cost of their meal.

```python
# plot a scatter plot and check the relationship between the variables 'tip' and 'bill amount'
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Create a scatter plot using Seaborn
sns.scatterplot(x='total_bill', y='tip', data=df)

# Add labels and a title
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")
plt.title("Scatter Plot of Tip Amount vs Total Bill Amount")

# Show the plot
plt.show()
```

#### **Q1. What type of Correlation do you see from the above Scatter plot? - 1 point**

Ans: From the scatter plot of 'tip' amount vs. 'total_bill' amount, it appears to show a positive correlation. This means that as the total bill amount increases, the tip amount also tends to increase.

<br>The strip plot resembles a scatterplot when one variable is categorical. This plot can help study the underlying distribution.


**2. Average tip amount by time of day:**

You can use a bar plot/strip plot to visualize the average tip amount by time of day (lunch or dinner). This will help you understand if people tip differently based on the time of day.

```python
# Plot a strip plot to check the relationship between the variables 'tip' and 'day'
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Create a strip plot using Seaborn
sns.stripplot(x='day', y='tip', data=df)

# Add labels and a title
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount")
plt.title("Strip Plot of Tip Amount vs Day of the Week")

# Show the plot
plt.show()
```

#### **Q2. What day of the week usually collects higher tip amounts? - 1 point**

Ans: Based on the strip plot, it appears that Saturday (day 6) and Sunday (day 7) of the week usually collect higher tip amounts compared to the other days.

```python
# Plot a strip plot to check the relationship between the variables 'tip' and 'time'
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Create a strip plot using Seaborn
sns.stripplot(x='time', y='tip', data=df)

# Add labels and a title
plt.xlabel("Time of Day")
plt.ylabel("Tip Amount")
plt.title("Strip Plot of Tip Amount vs Time of Day")

# Show the plot
plt.show()
```

```python
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Set up a grid of plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 3: Average tip amount by day of the week (Bar Plot)
sns.barplot(x='day', y='tip', data=df, ax=axes[0, 0])
axes[0, 0].set_xlabel("Day of the Week")
axes[0, 0].set_ylabel("Average Tip Amount")
axes[0, 0].set_title("Average Tip Amount by Day of the Week")

# Plot 4: Average tip amount by size of the party (Bar Plot)
sns.barplot(x='size', y='tip', data=df, ax=axes[0, 1])
axes[0, 1].set_xlabel("Party Size")
axes[0, 1].set_ylabel("Average Tip Amount")
axes[0, 1].set_title("Average Tip Amount by Party Size")

# Plot 5: Average tip amount by smoker status (Bar Plot)
sns.barplot(x='smoker', y='tip', data=df, ax=axes[1, 0])
axes[1, 0].set_xlabel("Smoker Status")
axes[1, 0].set_ylabel("Average Tip Amount")
axes[1, 0].set_title("Average Tip Amount by Smoker Status")

# Plot 6: Average tip amount by gender (Bar Plot)
sns.barplot(x='sex', y='tip', data=df, ax=axes[1, 1])
axes[1, 1].set_xlabel("Gender")
axes[1, 1].set_ylabel("Average Tip Amount")
axes[1, 1].set_title("Average Tip Amount by Gender")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

### 1.3 Distribution Plot

<br>A seaborn provides a `displot()` function which is used to visualize a distribution of the univariate variable. This function uses matplotlib to plot a histogram and fit a kernel density estimate (KDE)

```python
# plot a distribution plot of 'total_bill'
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Create a distribution plot for 'total_bill' using displot
sns.displot(df['total_bill'], kde=False, bins=20)

# Add labels and a title
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.title("Distribution Plot of Total Bill Amount")

# Show the plot
plt.show()
```

#### **Q6. What kind of skew does this plot have? - 1 point**

Ans: The distribution plot of 'total_bill' appears to have a right-skewed (positively skewed) distribution. In a right-skewed distribution, the tail of the distribution is stretched out towards the right, and the majority of the data points are concentrated on the left side. This suggests that there are relatively fewer instances of high total bill amounts compared to lower bill amounts in the dataset.

### 1.4 Count Plot

<br>Count plot shows the count of observations in each category of a categorical variable. We can add another variable using a parameter 'hue'.

```python
# Plot a countplot for count of observations by day
# set 'time' as hue parameter
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Create a countplot with 'day' on the x-axis and 'time' as the hue
sns.countplot(x='day', hue='time', data=df)

# Add labels and a title
plt.xlabel("Day of the Week")
plt.ylabel("Count of Observations")
plt.title("Count of Observations by Day with 'Time' Hue")

# Show the plot
plt.legend(title="Time of Day")  # Add a legend for 'time'
plt.show()
```

### 1.5 Heatmap

<br> Heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors. Each square in the heatmap shows the correlation between variables on each axis.

Compute correlation between the variables using `.corr()` function.

```python
# Plot a heatmap of the correlation matrix.
# Calculate the correlation matrix

# Load the "tips" dataset
df = sns.load_dataset('tips')

# Calculate the correlation matrix for the loaded dataset
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")

# Add a title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```

#### **Q8. Plot the correlation heatmap - 1 point**

```python
# plot heatmap
# 'annot=True' returns the correlation values, set cmap = 'YlGnBu'
# Load the "tips" dataset
df = sns.load_dataset('tips')

# Calculate the correlation matrix for the numeric columns
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")

# Add a title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```

### 2.3 Pie Plot

Pie plot is a graphical representation of univariate data. It is a circular graph divided into slices displaying the numerical proportion. For the categorical variable, each slice of the pie plot corresponds to each of the categories.



```python
# categories in the 'day' variable
df['day'].value_counts()
```

```python
# plot the occurrence of different days in the dataset

# 'autopct' displays the percentage upto 1 decimal place
# 'radius' sets the radius of the pie plot
plt.pie(df.day.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = ['Sat', 'Sun','Thur','Fri'])


# display the plot
plt.show()
```

**Observations**
* From the above pie plot, it can be seen that the data has a high proportion for Saturday followed by Sunday.

**Exploded pie plot** is a plot in which one or more sectors are separated from the disc. Use the `explode` argument to seperate the friday portion. Remeber to use `plt.show()`

#### **Q.10 Plot an exploding pie chart - 1 point**

```python
# plot the occurrence of different days in the dataset
# 'autopct' displays the percentage upto 1 decimal place
# 'radius' sets the radius of the pie plot


# display the plot
```

```python
# Get the count of each day in the dataset
day_counts = df['day'].value_counts()

# Define labels in the original order
labels = day_counts.index

# Create a list to specify which slice to explode
explode = [0.3 if label == 'Fri' else 0 for label in labels]

# Create a pie plot with 'autopct', 'radius', and 'explode' parameters
plt.pie(day_counts, autopct='%.1f%%', radius=1.2, labels=labels, explode=explode)

# Set an aspect ratio to make the pie circular
plt.axis('equal')

# Add a title
plt.title("Occurrence of Different Days")

# Display the plot
plt.show()
```

### 2.5 Box Plot
<br>Boxplot is a way to visualize the five-number summary of the variable. The five-number summary includes the numerical quantities like minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. Boxplot gives information about the outliers in the data. Detecting and removing outliers is one of the most important steps in exploratory data analysis. Boxplots also tells about the distribution of the data.


#### **Q12. Plot the boxplot of `'total_bill'` to check the distribution and presence of outliers in the variable. - 1 point**

```python
# plot a distribution of total bill


# add labels for five number summary
plt.text(x = 1.1, y = df['total_bill'].min(), s ='min')
# plt.text(x = 1.1, y = df.total_bill.quantile(0.25), s ='Q1')
# plt.text(x = 1.1, y = df['total_bill'].median(), s ='meadian (Q2)')
# plt.text(x = 1.1, y = df.total_bill.quantile(0.75), s ='Q3')
# plt.text(x = 1.1, y = df['total_bill'].max(), s ='max')


# add the graph title and axes labels

# display the plot

# Calculate the five-number summary
data_min = df['total_bill'].min()
Q1 = np.percentile(df['total_bill'], 25)
median = np.median(df['total_bill'])
Q3 = np.percentile(df['total_bill'], 75)
data_max = df['total_bill'].max()

# Plot the histogram
plt.hist(df['total_bill'], bins=50, color='blue', alpha=0.7)

# Add text annotations for the five-number summary
plt.text(x = 1.1, y = data_min, s ='min: {:.2f}'.format(data_min))
plt.text(x = 1.1, y = Q1, s ='Q1: {:.2f}'.format(Q1))
plt.text(x = 1.1, y = median, s ='median: {:.2f}'.format(median))
plt.text(x = 1.1, y = Q3, s ='Q3: {:.2f}'.format(Q3))
plt.text(x = 1.1, y = data_max, s ='max: {:.2f}'.format(data_max))

# Add the graph title and axes labels
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
